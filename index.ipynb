## Display the names of all the employees in Boston 

Hint: join the employees and offices tables.

```python
# Your code here

cur.execute("""SELECT lastName, firstName 
               FROM employees
               JOIN offices
               USING (officeCode)
               WHERE city = 'Boston'
               ;""")
df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

## Are there any offices that have zero employees?
Hint: Combine the employees and offices tables and use a group by.

```python
# Your code here

cur.execute("""SELECT offices.officeCode, productNameemployees.employeeNumber) 
               FROM offices
               JOIN employees
               USING (officeCode)
               GROUP BY officeCode 
               ORDER BY COUNT(employees.employeeNumber)
               ;""")
df = pd.DataFrame(cur.fetchall())
df.columns = [x[0] for x in cur.description]
df
```

## Level Up: Display the names of every individual product that each employee has sold

```python
# Your code here

cur.execute("""SELECT firstName, lastName, productName
               FROM employees e
               JOIN customers c                   
               ON e.employeeNumber = c.salesRepEmployeeNumber
               JOIN orders o 
               USING(customerNumber)
               JOIN orderdetails od 
               USING(orderNumber)                    
               JOIN products p 
               USING(productCode)
               GROUP BY lastName
               ORDER BY firstName""")
df = pd.DataFrame(cur.fetchall())
df.columns = [i[0] for i in cur.description]
print(len(df))
df
```

## Level Up: Display the Number of Products each employee has sold

```python
# Your code here

cur.execute("""SELECT orders.customerNumber, COUNT (orderdetails.quantityOrdered) 
               FROM orders
               JOIN orderdetails
               USING (orderNumber)
               GROUP BY customerNumber 
               ;""")
df1 = pd.DataFrame(cur.fetchall())
df1.columns = [x[0] for x in cur.description]
df1
```

```python
cur.execute("""SELECT employees.employeeNumber, customers.customerNumber 
               FROM employees
               LEFT JOIN customers
               ON employees.employeeNumber = customers.salesRepEmployeeNumber
               GROUP BY employeeNumber 
               ;""")
df2 = pd.DataFrame(cur.fetchall())
df2.columns = [x[0] for x in cur.description]
df2
```

```python
df3 = df2.merge(df1,
               left_on='customerNumber',
               right_on='customerNumber',
               how='outer')
```

```python
df3.sort_values(['COUNT (orderdetails.quantityOrdered)'], ascending=[False])
```
